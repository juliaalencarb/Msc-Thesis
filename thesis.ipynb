```python
import math as mt
import numpy as np
import pandas as pds
import matplotlib
import matplotlib.pyplot as plt
```

```python
# CHART -> Figure 3 - ATP levels positively correlate with the number of living cells
df = pds.read_csv('./Data/Figure_3_Data.csv')

fig = plt.figure()
ax = plt.gca()
ax.scatter(df['Cells'], df['Mean'], c='Black', s=20)

z = np.polyfit(df['Cells'], df['Mean'], 1)
p = np.poly1d(z)

ax.plot(df['Cells'], p(df['Cells']), 'k--')
ax.errorbar(df['Cells'], df['Mean'], yerr=df['SE'], linewidth=1, capsize=3, color='Black')

plt.xlabel(r'Number of cells (x$10^3$)', fontsize=10, fontweight='bold')
plt.ylabel(r'Luminescence (x$10^3$)', fontsize=10, fontweight='bold')

plt.show()
```

```python
# CHART -> FIGURE 4 - Effects of NVP-BEZ235 on cell viability after treatment
df = pds.read_csv('./Data/Figure_4_Data.csv')

ax = df.plot.scatter('Concentration', 'Mean', c='Black', s=20)
ax.errorbar(df['Concentration'], df['Mean'], yerr=df['SE'], linewidth=1, capsize=3, color='Black')
ax.set_xscale('log')

plt.text(df['Concentration'][1], df['Mean'][1] + df['SE'][1], "*", ha='center', va='bottom', color='k')
plt.text(df['Concentration'][2], df['Mean'][2] + df['SE'][2], "*", ha='center', va='bottom', color='k')
plt.text(df['Concentration'][3], df['Mean'][3] + df['SE'][3], "*", ha='center', va='bottom', color='k')
plt.text(df['Concentration'][4], df['Mean'][4] + df['SE'][4], "*", ha='center', va='bottom', color='k')

plt.xlabel('NVP-BEZ235 concentration (μM)', fontsize=10, fontweight='bold')
plt.ylabel('Cell Viability', fontsize=10, fontweight='bold')

ax.set_xlim(0.001, 10)
ax.get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter()) 
ax.get_xaxis().set_tick_params(which='minor', size=0)
ax.get_xaxis().set_tick_params(which='minor', width=0)

# Fixing first tick
a = ax.get_xticks()
a[0] = 0.001
ax.set_xticklabels(a)

plt.show()
```

```python
# CHART -> Figure 5 - Effects of NVP-BEZ235 on cell viability after treatment
df = pds.read_csv('./Data/Figure_5_Data.csv')

x = np.linspace(0, 1.2, 7)
colors = ['darkgrey', 'gainsboro']

fig = plt.figure()
ax = plt.gca()
ax.bar(df['Item'], df['Values'], width=0.7, align='center', edgecolor='Black', color=colors)
ax.errorbar(df['Item'], df['Values'], fmt='none', yerr=df['SE'], elinewidth=1, linewidth=1, capsize=3, color='Black')

plt.text(df['Item'][1], df['Values'][1] + df['SE'][1], "*", ha='center', va='bottom', color='k')
plt.xlabel('Treatment', fontsize=10, fontweight='bold')
plt.ylabel('Cell Viability (48 hours)', fontsize=10, fontweight='bold')

plt.show()
```
